```python
def calculate_brightness(img) -> float:
    """
    Calculate the average brightness of a grayscale image using PyTorch.

    Args:
        img: A 2D list where each element represents a pixel value between 0-255.

    Returns:
        The average brightness rounded to two decimal places, or -1 for invalid inputs.
    """
    #edge case number 1 : img is empty 
    if (img == []):
        return -1
    

    #edge case number 2: if the matrix is not a 2D array 
    n_columns = len(img)
    ref = len(img[0])
    k=0
    while (k<n_columns and ref==len(img[k])):
        k+=1
    if (k<n_columns):
        return -1
    
    n_lines = len(img[0])
    b=True
    #edge case number 3: 
    for i in range(n_columns):
        for j in range(n_lines):
            b = b and (0<=img[i][j]<=255)
    if (not b):
        return -1
    
    # general case
    return np.asarray(img).sum()/(n_columns*n_lines)
```

```python
img = [
    [100, 200],
    [50, 150]
]
print(calculate_brightness(img))
```
